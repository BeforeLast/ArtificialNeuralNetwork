```python
import os
import pandas as pd
from classes.layers.LSTM import LSTM
from classes.models.Sequential import Sequential
from classes.layers.Input import InputLayer
from classes.layers.Dense import Dense
```

# Data and Preprocessing

```python
df = pd.read_csv("data/ETH-USD/ETH-USD-Test.csv")

# Drop Date Column 
df = df.drop("Date", axis=1)

# Take Last 32 data
df = df.tail(32)

df
```

## Data Shape

```python
data_shape = df.shape
data_shape
```

# Model

```python
model = Sequential([
  InputLayer(input_shape=data_shape),
  LSTM(32),
  Dense(5)
])
model.compile()
```

```python
output = model.predict(df.values)
print(output)
```

# Output Analysis

Dari output tersebut, dapat disimpulkan bahwa bahwa prediksi nilai selanjutnya dari kolom Open, High, Low, Close, dan Volume dibulatkan ke tiga angka di belakang koma berturut-turut adalah 17,409, 15,864, 15,755, 17,540, dan 15,522. Tentunya hasil ini memiliki perbedaan yang signifikan dengan data-data sebelumnya. Hal ini dikarenakan model yang digunakan untuk memprediksi output belum di-train dan hanya menggunakan inisialisasi bobot secara acak. Kedepannya, apabila model ini dilakukan proses training, maka diharapkan output yang dihasilkan akan menjadi lebih baik dan relevan dengan data-data sebelumnya, sehingga dapat memberikan prediksi yang relatif lebih akurat.
